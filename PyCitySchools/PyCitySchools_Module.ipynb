```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "../Resources/schools_complete.csv"
student_data_to_load = "../Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete
```

```python
# Calculate total number of schools and students using a len statement statement in pandas
School_count = len(school_data_complete["school_name"].unique())
Student_count = len(school_data_complete["student_name"])
budget = school_data_complete["budget"].unique()
sum_budget = budget.sum()
```

```python
math_average = school_data_complete["math_score"].mean()
reading_average = school_data_complete["reading_score"].mean()
reading_average
```

```python
passing_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
reading_passing = (len(passing_reading)/Student_count)*100
passing_math = school_data_complete.loc[school_data_complete["math_score"] >= 70]
math_passing = (len(passing_math)/Student_count)*100
passing = school_data_complete.loc[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70),:]
passing_both=(len(passing)/Student_count)*100
```

```python
summary_df = pd.DataFrame({"Total Schools": [School_count],
                              "Total Students": [Student_count],
                              "Total Budget": [sum_budget],
                              "Average Math Score": [math_average],
                            "Average Reading Score": [reading_average],
                              "% Passing Math": [math_passing],
                              "% Passing Reading": [reading_passing],
                              "% Passsing Both": [passing_both]})
summary_df
```

```python
school = school_data_complete.groupby(["school_name"])
Type_count = school["type"].first()
# create variable that gives total students per school
Total_students = school["Student ID"].count()
#create variable that gives Total School Budget per school
budget_total = school["budget"].first(0)
#create variable that gives $ per student at specified school
budget_per_student = budget_total/Total_students
#create average math score per student at specified school
average_math = school["math_score"].mean()
#create average reading score per student at specified school
average_reading = school["reading_score"].mean()

#create % passing for math
pct_passingmath  = school_data_complete.loc[school_data_complete["math_score"] >= 70]
passing_math = pct_passingmath.groupby(["school_name"])
math = 100*(passing_math["math_score"].count()/Total_students)

#create % passing for reading
pct_passingreading  = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
passing_reading = pct_passingreading.groupby(["school_name"])
reading = 100*(passing_reading["reading_score"].count()/Total_students)

# Create Overall Passing %(both math and reading)
pct_passing = school_data_complete.loc[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70),:]
passing_overall = pct_passing.groupby(["school_name"])
overall = 100*(passing_overall["reading_score"].count()/Total_students)



#Create a data frame including all variables stated above 
total_summary_df = pd.DataFrame({"School Type": Type_count,
                              "Total Students": Total_students,
                              "Total School Budget": budget_total,
                              "Per Student Budget": budget_per_student,
                            "Average Math Score": average_math,
                              "Average Reading Score": average_reading,
                              "% Passing Math": math,
                              "% Passsing Reading": reading,
                               "% Overall Passing": overall})
total_summary_df
```

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labelsy = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
school_spending_df = total_summary_df.copy()
```

```python
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], spending_bins, labels=labelsy,include_lowest = True)
school_spending_df.head()
```

```python
spending_df = school_spending_df.groupby(["Spending Ranges (Per Student)"])
spavgmath = spending_df["Average Math Score"].mean()
spavgreading = spending_df["Average Reading Score"].mean()
spavgmathpct = spending_df["% Passing Math"].mean()
spavgreadingpct = spending_df["% Passsing Reading"].mean()
spavgoverallpct = spending_df["% Overall Passing"].mean()

spavgoverallpct
```

```python
school_range_summary = pd.DataFrame({"Average Math Score": spavgmath,
                              "Average Reading Score": spavgreading,
                              "% Passing Math": spavgmathpct,
                              "% Passsing Reading": spavgreadingpct,
                               "% Overall Passing": spavgoverallpct})
school_range_summary
```
